```python
# specifying the cloumn names I want to use
new_cols_dict = {
    'timestamp': 'time',
    'cnt': 'count',
    't1': 'temp_real_C',
    't2': 'temp_feels_like_C',
    'hum': 'humidity_percent',
    'wind_speed': 'wind_speed_kph',
    'weather_code': 'weather',
    'is_holiday': 'is_holiday',
    'is_weekend': 'is_weekend',
    'season': 'season'
}

bikes_df.rename(new_cols_dict, axis=1, inplace=True)
bikes_df
```

```python
# changing the humidity to percentage
bikes_df.humidity_percent = bikes_df.humidity_percent/100
bikes_df.head()
```

```python
bikes_df['season'].dtype
```

```python
bikes_df['weather'].dtype
```

```python
# creating a mapping dictionary for seasons and weather
season_dict = {
    '0.0': "spring",
    '1.0': "summer",
    '2.0': "autumn",
    '3.0': "winter",
}

weather_dict = {
    "1.0": "Clear",
    '2.0': "Scattered Clouds",
    '3.0': "Broken Clouds",
    '4.0': "Cloudy",
    '7.0': "Rain",
    '10.0': "Rain with thunderstorm",
    '26.0': "Snowfall"
}

# changing season column data type to string
bikes_df.season = bikes_df.season.astype('str')
bikes_df.weather = bikes_df.weather.astype('str')
```

```python
bikes_df.info()
```

```python
# mapping the values
bikes_df.season = bikes_df.season.map(season_dict)
bikes_df.weather = bikes_df.weather.map(weather_dict)
```
